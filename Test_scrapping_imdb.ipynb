```python
from concurrent.futures import ThreadPoolExecutor
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import json
import re

df_actors = pd.read_csv('ID_Actors.csv', sep=';')

start_time = time.time()

HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
for i in df_actors['ID_Star1'].iloc[:100]:
    print("Pour", i, "--", df_actors['Name_Star1'].loc[df_actors['ID_Star1'] == i].values[0])
    url = f'https://www.imdb.com/name/{i}/'
    session = requests.Session()
    res = session.get(url, headers=HEADERS) 

    soup = BeautifulSoup(res.content, 'lxml')

    script = soup.find('script', {'id': '__NEXT_DATA__', 'type': 'application/json'})

    if script:
        script_content = script.string
        data_dict = json.loads(script_content)

    if 'props' in data_dict:
        props_dict = data_dict['props']

        # Vérifiez si 'pageProps' existe dans le dictionnaire 'props'
        if 'pageProps' in props_dict:
            page_props_dict = props_dict['pageProps']

            # Vérifiez si 'aboveTheFold' existe dans le dictionnaire 'pageProps'
            if 'aboveTheFold' in page_props_dict:
                above_the_fold_dict = page_props_dict['aboveTheFold']

                # Vérifiez si 'meterRanking' existe dans le dictionnaire 'aboveTheFold'
                if 'meterRanking' in above_the_fold_dict:
                    meter_ranking_dict = above_the_fold_dict['meterRanking']
                    if meter_ranking_dict:
                        # Vérifiez si 'currentRank' existe dans le dictionnaire 'meterRanking'
                        if 'currentRank' in meter_ranking_dict:
                            current_rank = meter_ranking_dict['currentRank']
                            print("La valeur de 'currentRank' est :", current_rank)
                        else:
                            print("'currentRank' n'a pas été trouvé dans le dictionnaire 'meterRanking'.")
                    else:
                        print("Pas de meterranking")
                else:
                    print("'meterRanking' n'a pas été trouvé dans le dictionnaire 'aboveTheFold'.")
            else:
                print("'aboveTheFold' n'a pas été trouvé dans le dictionnaire 'pageProps'.")
        else:
            print("'pageProps' n'a pas été trouvé dans le dictionnaire 'props'.")
    else:
        print("'props' n'a pas été trouvé dans le JSON principal.")

else:
    print("Script not found.")

```
