```python
# Package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

allocine = pd.read_excel(r"C:\Users\maoui\OneDrive\Documents\Data Scientest\DTS - Projet\allocineV3-ND.xlsx")
allocine.head()
```

```python
# Créer un nouveau DataFrame avec les colonnes ID et genres
allocine_genres = allocine.loc[:, ['ID', 'genres']]

# Afficher les premières lignes du nouveau DataFrame pour vérification
print(allocine_genres.head())
```

```python
# Liste des genres
genres_list = ['Action', 'Animation', 'Aventure', 'Biopic', 'Comedie', 'Comedie dramatique', 'Comedie musicale', 'Drame', 'Epouvante-horreur', 'Famille', 'Fantastique', 'Guerre', 'Historique', 'Musical', 'Policier', 'Romance', 'Science Fiction', 'Thriller', 'Western']

# Créer un DataFrame temporaire avec des variables indicatrices pour chaque genre
genre_dummies = allocine_genres['genres'].str.get_dummies(sep=', ')

# Sélectionner uniquement les colonnes correspondant aux genres de la liste
genre_dummies = genre_dummies[genres_list]

# Fusionner le DataFrame des variables indicatrices avec allocine_genres sur la colonne 'ID'
allocine_genres = pd.concat([allocine_genres['ID'], genre_dummies], axis=1)

allocine_genres.head()
```

```python
# Enregistrer le DataFrame au format CSV
allocine_genres.to_csv(r'C:\Users\maoui\OneDrive\Documents\Data Scientest\DTS - Projet\allocine_genres.csv', index=False)
```
