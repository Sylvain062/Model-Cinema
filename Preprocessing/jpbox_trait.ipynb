```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
import re
import unicodedata
from unidecode import unidecode
os.chdir('D:\Sylvain\GitHub\Model-Cinema')
#***********************************************************************
df=pd.read_csv('Data/jpbox new devise.csv',sep=';')
df.head()
# 'on enlève les accents des titres'#
df['titre']=df['titre'].str.replace(' ', '', regex=True) # on enlève tous les espaces
df['titre'].fillna('XXXX', inplace=True) # ca sert à éviter de faire planter le split si valeur Nan
df['titre']=df['titre'].apply(unidecode) #on enlève les accents
df['titre']=df['titre'].replace('XXXX',np.NaN,regex=True) # et je retire les XXXX
df.head()

df2=pd.read_csv('Data/allocineV8-SB.csv')
df2.head(20)
# 'on enlève les accents des titres'#
df2['titre_original']=df2['titre_original'].str.replace(' ', '', regex=True) # on enlève tous les espaces
df2['titre_original'].fillna('XXXX', inplace=True) # ca sert à éviter de faire planter le split si valeur Nan
df2['titre_original']=df2['titre_original'].apply(unidecode) #on enlève les accents
df2['titre_original']=df2['titre_original'].replace('XXXX',np.NaN,regex=True) # et je retire les XXXX
df2.head()

# on fait un join left sur le titre
merge = pd.merge(df2, df, left_on = 'titre_original', right_on = 'titre',how='left')
merge.head(20)
merge.to_csv('Data/merge.csv')
df2.head(20)
```
